```python
from selenium import webdriver # driver가동
from selenium.webdriver.chrome.options import Options # binary 잡아주기 위한
from selenium.webdriver.common.keys import Keys # 텍스트 값 보내기 위한 모듈
from selenium.webdriver.support.ui import WebDriverWait # 웹페이지 기다리기
#time.sleep() 을 줄수도 잇지만 언제 까지 기다려야 하는 지몰라서 
#WebDriverWait 엘리먼트가 나올떄 까지마 기다리는것.
from selenium.webdriver.common.by import By
from selenium.webdriver.support import expected_conditions as EC

import requests
from bs4 import BeautifulSoup
import time
import pandas as pd
```

```python
# broswer = webdriver.Chrome('./chromedriver.exe',chrome_options = options) # 괄호안에 경로 설정 똑같은 경로면 따로 안해도됨 !
#위 코드 실행시 다음과 같은 오류 
#cannot find Chrome binary 
#option을 통한 binary경로를 잡아주기

options = Options() 
options.binary_location= 'C:/Program Files/Google/Chrome/Application/chrome.exe'
broswer = webdriver.Chrome(chrome_options = options)
```

```python
broswer.get('https://play.google.com/store/apps/details?id=gov.cheongju.cjpay&showAllReviews=true') # 셀레늄을 통해 열 브라우저 선택 
```

```python
interval=2
broswer.maximize_window() # 브라우저 최대화 
#현재 문서 높이를 저장 
prev_height = broswer.execute_script('return document.body.scrollHeight')
#반복 수행 

while True:
    #스크롤을 가장 아래로 내림 
    broswer.execute_script('window.scrollTo(0,document.body.scrollHeight)')
    #페이지 로딩 대기 
    time.sleep(interval)
    #현재 문서 높이를 가져와서 저장 
    curr_height = broswer.execute_script('return document.body.scrollHeight')
    if curr_height ==prev_height:
        break
    prev_height = curr_height
print('스크롤 완료')
```

```python
soup = BeautifulSoup(broswer.page_source,'lxml') # 셀레늄으로 한 페이지 소스 

```

```python
###########리뷰 ##############3
#리뷰 리스트 생성 162개 
lst_review=[]
reviews = soup.find_all('span',attrs={'jsname':'bN97Pc'})
for review in reviews:
    review = review.get_text()
    lst_review.append(review)
    print(review)
```

```python
###############이정보가 도움이 되요 #######################33
#좋아요 리스트 생성 162개 
lst_goods = []
goods = soup.find_all('div',attrs={'class':'jUL89d y92BAb'})
for good in goods:
    like = good.get_text()
    lst_goods.append(like)
    print(like)
print(len(lst_goods))
```

```python
#별점 데이터를 수집할 것인데, 그냥 별점 데이터를 수집하면 가장 상단의 평균 별점까지 수집을 해버림.
#frame라는 객체 생성을 통해 별점을 수집할 항목을 지정해 놓고 수집 
frame = soup.find('div',attrs={'jsname':'lYU69'})
print(frame) # 
```

```python
######################별표 평점 #########################
stars = frame.find_all('div',attrs={'class':'pf5lIe'}) # 162개
lst_stars =[]
for star in stars:
    num_star = star.find('div').attrs['aria-label']
    lst_stars.append(num_star)
    print(num_star)
#162개 
```

```python
#################날짜####################################
frame_dates = soup.find_all('div',attrs={'class':'xKpxId zc7KVe'})
lst_date=[]
for date in frame_dates:
    date = date.find('span',attrs={'class':'p2TkOb'})
    date = date.get_text()
    lst_date.append(date)
    print(date)
print(len(lst_date))
```
