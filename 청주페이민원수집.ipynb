```python
broswer.get('https://siseon.cheongju.go.kr/homepage/research/square/313') # 셀레늄을 통해 열 브라우저 선택 
```

```python
i=0
while True:
    try:
        elem = broswer.find_element_by_xpath('//*[@id="app"]/div[2]/div[2]/div[2]/span/a')
        elem.click()
        time.sleep(3)
        i +=1
        print(i)
    except NoSuchElementException:
        print('end')
        
        

    #여기를 어떻게 해야 멋지게 마무리 할까 ... try execpt?
```

```python
soup = BeautifulSoup(broswer.page_source,'lxml')
soup
```

```python
reviews = soup.find_all('p',attrs={'style':'white-space: pre-wrap;'})
lst_review=[]
for review in reviews:
    review = review.get_text()
    lst_review.append(review)
lst_review
 
```

```python
print(len(lst_review))
```

```python
dates = soup.find_all('span')
lst_date =[]
for date in dates:
    date = date.get_text()
    lst_date.append(date)
    
    
```

```python
lst_date[4]
```

```python
lst = []
for i in range(len(lst_date)):
    if i==0:
        continue
    elif i%2 ==1:
        lst.append(lst_date[i])
lst
```

```python
lst = lst[3:]
lst
```

```python
df_cheong = pd.DataFrame({'content':lst_review
             ,'date':lst})
```
